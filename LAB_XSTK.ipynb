```r
install.packages("patchwork", repos = "http://cran.us.r-project.org")
library(ggplot2)
library(patchwork)

plot_sequence <- function(n_value) {
  n <- 1:n_value
  x_n <- (1 + 1/n)^n
  e <- exp(1)

  df <- data.frame(n, x_n)

  ggplot(df, aes(x = n, y = x_n)) +
    geom_line(color = "steelblue", size = 1) +
    geom_hline(yintercept = e, linetype = "dashed", color = "red") +
    labs(
      x = "n",
      y = expression(x[n])
    ) +
    theme_minimal(base_size = 15)
}

p1 <- plot_sequence(10)
p2 <- plot_sequence(50)
p3 <- plot_sequence(100)
p4 <- plot_sequence(500)
p5 <- plot_sequence(1000)

final_plot <- (p1 | p2) / (p3 | p4) / p5 +
  plot_annotation(title = "Hoi tu cua (1 + 1/n)^n → e voi cac kich thuoc mau khac nhau")

print(final_plot)
```

```r
library(gridExtra)
library(ggplot2)

n_sim <- 10000
sample_sizes <- c(1, 2, 5, 10, 30 ,100)

clt_sim <- function(n) {
  samples <- matrix(rexp(n_sim * n, rate = 1), ncol = n)
  sums <- rowSums(samples)
  means <- sums / n
  standardized <- (sums - n * 1) / sqrt(n * 1)

  list(sum = sums, mean = means, z = standardized)
}

sim_data <- lapply(sample_sizes, clt_sim)
names(sim_data) <- paste0("n", sample_sizes)

plots <- list()

for (i in seq_along(sample_sizes)) {
  n <- sample_sizes[i]
  z <- sim_data[[i]]$z
  df <- data.frame(z = z)

  p <- ggplot(df, aes(x = z)) +
    geom_histogram(aes(y = after_stat(density)), bins = 50,
  fill = "skyblue", alpha = 0.7, color = "black") +
    stat_function(fun = dnorm, color = "red", size = 1.2, linetype = "dashed") +
    labs(title = paste("n =", n),
         subtitle = "Chuan hoa Zn",
         x = "Gia tri chuan hoa", y = "Mat do")

  plots[[i]] <- p
}

grid.arrange(grobs = plots, ncol = 3)
```

```r
# Ôn thi

n <- 1000
p_true <- 0.52

x_supporters <- rbinom(1, size = n, prob = p_true)

p_mu <- x_supporters / n

cat("So nguoi ung ho mau p_mu:", x_supporters, "\n")
cat("Ti le ung ho mau p_mu:", p_mu, "\n\n")

sai_so_chuan <- sqrt(p_mu * (1 - p_mu) / n)

z_score <- qnorm(0.975)

dung_sai <- z_score * sai_so_chuan

lower_bound <- p_mu - dung_sai
upper_bound <- p_mu + dung_sai

cat("Khoang tin cay 95% cua ti le mau:", round(lower_bound, 4), "-", round(upper_bound, 4), "\n")
```

```r
# Tạo ngẫu nhiên 35 giá trị của biến ngẫu nhiên có phân phối chuẩn với trung bình bằng
# 10 và độ lệch chuẩn 5. Tìm khoảng tin cậy 95% cho kỳ vọng của biến ngẫu nhiên chuẩn
# dựa vào số liệu vừa tạo.

n <- 35
muy <- 10
sigma <- 5

data <- rnorm(n, mean = muy, sd = sigma)

x_mean <- mean(data)
do_lech_chuan <- sd(data)

alpha = 0.05
epsilon <- qnorm(1 - alpha / 2) * do_lech_chuan / sqrt(n)

lower_bound = x_mean - epsilon
upper_bound = x_mean + epsilon

cat(round(lower_bound, 4), " - ", round(upper_bound, 4))
```
